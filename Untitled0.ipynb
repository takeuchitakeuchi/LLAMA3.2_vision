```python
# 4) モデル読み込み（4bit量子化で省メモリ）
# T4でも動くように4bitで読み込みます。A100ならbfloat16に変えてもOKです。

import torch
from transformers import MllamaForConditionalGeneration, AutoProcessor, BitsAndBytesConfig

MODEL_ID = "meta-llama/Llama-3.2-11B-Vision-Instruct"

# 省メモリ 4bit 量子化（T4想定）
quant = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.bfloat16,  # A100ならbfloat16が安定
)
model = MllamaForConditionalGeneration.from_pretrained(
    MODEL_ID,
    quantization_config=quant,
    device_map="auto",
)
processor = AutoProcessor.from_pretrained(MODEL_ID)

print("Loaded on:", model.device)
```

```python
# 5) フォルダ内の画像に対して「手すりの有無」を質問
# 画像ごとに はい/いいえ + 一言理由 で日本語回答させます。
# 誤検出を減らすため、ガードレール/手すりの例を英語で補助定義しています（視覚と言語の安定化のため）。


import glob
from PIL import Image
from pathlib import Path

# 画像フォルダ（必要に応じて変更）
image_dir = Path("/content/drive/MyDrive/code/LLAMA3_2_vision_11B/llama_images")
image_paths = sorted(
    list(image_dir.glob("*.jpg")) +
    list(image_dir.glob("*.jpeg")) +
    list(image_dir.glob("*.png")) +
    list(image_dir.glob("*.JPG")) +
    list(image_dir.glob("*.PNG"))
)

assert len(image_paths) > 0, f"No images found in {image_dir}. Put images there."

SYSTEM_INSTRUCT = (
    "You are a safety inspector. Determine if the image contains a construction-site handrail "
    "(a guardrail consisting of top/mid rails and posts near edges/openings, typically metal or wood). "
    "Ignore road guardrails if they are far and irrelevant to a work platform. "
    "Answer in Japanese with 'はい' or 'いいえ' first, then one short reason."
)

def ask_handrail(img: Image.Image) -> str:
    # chatテンプレート: 画像+テキストの会話形式
    messages = [
        {"role": "system", "content": SYSTEM_INSTRUCT},
        {"role": "user", "content": [
            {"type": "image"},
            # {"type": "text", "text": "Are there any 1)handrails, 2)handrails at the middle,  3)base boards in this image? Judge sepalately"}
            {"type": "text", "text": "explain the image"}
        ]},
    ]
    prompt = processor.apply_chat_template(messages, add_generation_prompt=True)
    inputs = processor(img, prompt, add_special_tokens=False, return_tensors="pt").to(model.device)

    with torch.inference_mode():
        output = model.generate(
            **inputs,
            max_new_tokens=96,
            do_sample=False,
            temperature=None,
        )
    text = processor.decode(output[0], skip_special_tokens=True)
    return text.strip()

results = []
for p in image_paths:
    try:
        img = Image.open(p).convert("RGB")
        ans = ask_handrail(img)
        print(f"[{p.name}] -> {ans}")
        results.append((p.name, ans))
    except Exception as e:
        print(f"[{p.name}] ERROR: {e}")
```

```python
# 6)（任意）結果をCSVに保存

import csv
out_csv = "/content/handrail_results.csv"
with open(out_csv, "w", newline="", encoding="utf-8") as f:
    w = csv.writer(f)
    w.writerow(["filename", "answer"])
    w.writerows(results)
print("Saved:", out_csv)

```
